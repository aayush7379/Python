```python
# image drawing
from PIL import ImageFont
img = im1.resize((w//2,h//2))
dr = ImageDraw.Draw(img)
start_point = (0,img.height//2)
end_point = (img.width, img.height//2)
dr.line(xy=(start_point,end_point),fill='white',width=5)
dr.line(xy=(img.width//2,0),fill='white',width=5)
dr.circle((180,320),radius =100 , fill = 'red')
dr.circle((180,320),radius =100 , outline='yellow',width=40)
font = ImageFont.truetype(r"c:\WINDOWS\Fonts\ALGER.TTF",50)
dr.text((150,300),"Target",fill="black",font=font,
         stroke_fill='white', stroke_width=2,anchor='ms')
img

```
