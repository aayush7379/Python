# boolean indexing
`df['column'] > value`

`df['column'] == value`

`df['column'] != value`

```python
df[df['price']>1000000]
```

```python
df[df['condition']>=5]
```

```python
df[df['condition']<=1]
```

```python
# 5 starts rated houses + 2 floors 
df[(df['condition']==5) & (df['floors']==2)]
```

```python
f1 = df['price']>1000000
f2 = df['condition'] == 5
f3 = df['floors'] == 2
df[f1 & f2 & f3]
```

```python
df[f1&f2&f3].reset_index().style.highlight_max(color='blue')
```

```python
df[f1&f2&f3].reset_index().style.background_gradient(cmap='GnBu')
```

```python
df[f1&f2&f3].reset_index().style.set_properties(color='red',subset=['price'])
```

```python
# query 
df.query('price > 2000000')
```

```python
df.query('price > 2000000 and condition == 5')
```

```python
df.query('price > 2000000 and condition == 5').sort_values('price')
```

```python
df.columns
```

```python
cols_to_drop = ['sqft_living15','sqft_lot15']
df = df.drop(columns=cols_to_drop)
df
```

```python
rows_to_drop = range(1000)
df.drop(index=rows_to_drop, axis=0)
```

```python
df.drop_duplicates(subset=['date']) # always use subsets column
```

```python
df.dropna() # remove eows with NaN values
```

```python
# renaming coloums 
df.columns
```

```python
name_update = {
    'lat':'lattitude',
    'long':'longitude'
}
df.rename(columns=name_update)
```

```python
# dummy column add 
df['dummy'] = "house_price" # repeat same value in all rows 
```

```python
df
```

```python
df['age'] = 2024 - df['yr_built']
df.head()
```
